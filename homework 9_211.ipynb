```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
exercise = pd.read_csv('C:/Users/chemi/OneDrive/Coding/Exercise_Data.csv')
exercise.head()
```

```python
# Select only numeric columns
numeric_cols = exercise.drop(columns=['id', 'diet', 'kind'])
numeric_cols = ((numeric_cols - numeric_cols.min()) / (numeric_cols.max() - numeric_cols.min()))

# Plot the heatmap
sns.heatmap(numeric_cols, cmap='coolwarm')
plt.title("Pulse Dataset Heatmap")
plt.xlabel('Duration')
plt.ylabel('Pulse')
plt.show()
```

```python
# Create a boxplot of 1 min by Diet
sns.boxplot(data=exercise, x='diet', y='1 min', hue = 'kind')
plt.title("Boxplot of 1 min Pulse by Diet and Exercise Kind")
```

```python
# Create a boxplot of 15 min by Diet
sns.boxplot(data=exercise, x='diet', y='15 min', hue = 'kind', palette='viridis')
plt.title("Boxplot of 15 min Pulse by Diet and Exercise Kind")
```

```python
# Create a boxplot of 30 min by Diet
sns.boxplot(data=exercise, x='diet', y='30 min', hue = 'kind', palette='winter')
plt.title("Boxplot of 30 min Pulse by Diet and Exercise Kind")
```

```python
#The boxplots of the elementary students shows those that have a no diet have a higher pulse rate than those that have a low fat diet. You can also tell that those that are running have a higher pulse rate than those that are walking or resting for both sets of diets
```

```python
#load planets dataset
planets = sns.load_dataset('planets')
print(planets)
```

```python
sns.relplot(data=planets, x = 'year', y ='distance', hue='method', palette='winter')
plt.title("Distance from Earth by Year and Method")
plt.xlabel('Year')
plt.ylabel('Distance')
plt.show()
```

```python
sns.scatterplot(data=planets, x = 'year', y ='method', palette='seismic')
plt.title("Method by Year")
plt.xlabel('Year')
plt.ylabel('Method')
plt.show()
```

```python
sns.histplot(data=planets, x='year', hue='method', palette='viridis')
plt.title("Year by Method") 
plt.xlabel('Year')
plt.ylabel('Method')
plt.show()
```

```python
sns.histplot(data=planets, x='distance', hue='method', palette='coolwarm')
plt.title("Distance by Method")
plt.xlabel('Distance')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(data=planets, x='method', y='year', palette='viridis')
plt.title("Method by Year")
plt.xlabel('Method')
plt.ylabel('Year')
plt.show()
```

```python
sns.boxplot(data=planets, x='method', y='distance', palette='Greens')
plt.title("Method by Distance")
plt.xlabel('Method')
plt.ylabel('Distance')
plt.show()
```

```python
# I feel the histogram of the year by method is the most informative because it shows the distribution of the methods used to find the planets over the years. It shows that the radial velocity method has been used the most over the years.
```
